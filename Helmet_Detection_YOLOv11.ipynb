```python
# Helmet Detection with YOLOv11 - Google Colab Script
# Complete setup for training and inference

# =============================================================================
# 1. INITIAL SETUP AND DEPENDENCIES
# =============================================================================

# Install required packages
!pip install ultralytics roboflow supervision opencv-python matplotlib pillow

# Import necessary libraries
import os
import yaml
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Image
from ultralytics import YOLO
import supervision as sv
import shutil

# =============================================================================
# 2. DATASET DOWNLOAD FROM ROBOFLOW
# =============================================================================

# Install and setup Roboflow
!pip install roboflow

from roboflow import Roboflow

# Initialize Roboflow with your API key
rf = Roboflow(api_key="0Bj5V6Y2QAIK6qrhAf2L")
project = rf.workspace("faiqa-eo2gt").project("helmet-and-license-plate-rwqee")
version = project.version(1)

# Download dataset in YOLOv11 format
print("Downloading dataset...")
dataset = version.download("yolov11")
print(f"Dataset downloaded to: {dataset.location}")

# =============================================================================
# 3. DATASET CONFIGURATION
# =============================================================================

# Get dataset path
dataset_path = dataset.location
data_yaml_path = os.path.join(dataset_path, "data.yaml")

# Read and display dataset configuration
with open(data_yaml_path, 'r') as file:
    data_config = yaml.safe_load(file)

print("Dataset Configuration:")
print(f"Train images: {data_config.get('train', 'N/A')}")
print(f"Validation images: {data_config.get('val', 'N/A')}")
print(f"Test images: {data_config.get('test', 'N/A')}")
print(f"Number of classes: {data_config.get('nc', 'N/A')}")
print(f"Class names: {data_config.get('names', 'N/A')}")

# =============================================================================
# 4. MODEL INITIALIZATION AND TRAINING
# =============================================================================

# Initialize YOLOv11 model (you can use yolov8n.pt, yolov8s.pt, yolov8m.pt, yolov8l.pt, yolov8x.pt)
model = YOLO('yolov8n.pt')  # Using nano version for faster training

# Training parameters
training_params = {
    'data': data_yaml_path,
    'epochs': 50,  # Adjust based on your needs
    'imgsz': 640,
    'batch': 16,   # Adjust based on your GPU memory
    'device': 0,   # Use GPU if available
    'workers': 2,
    'patience': 10,
    'save_period': 10,
    'project': 'helmet_detection',
    'name': 'train_run',
}

print("Starting training...")
print(f"Training parameters: {training_params}")

# Train the model
results = model.train(**training_params)

print("Training completed!")
print(f"Best model saved at: /content/helmet_detection/train_run/weights/best.pt")

# =============================================================================
# 5. MODEL EVALUATION
# =============================================================================

# Load the best trained model
best_model = YOLO('/content/helmet_detection/train_run/weights/best.pt')

# Validate the model
val_results = best_model.val(data=data_yaml_path)

print("Validation Results:")
print(f"mAP50: {val_results.box.map50}")
print(f"mAP50-95: {val_results.box.map}")

# =============================================================================
# 6. INFERENCE FUNCTIONS
# =============================================================================

def predict_image(model, image_path, conf_threshold=0.5):
    """
    Predict helmet detection on a single image
    """
    results = model(image_path, conf=conf_threshold)
    return results

def visualize_predictions(image_path, results, save_path=None):
    """
    Visualize predictions on an image
    """
    # Read the image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Get predictions
    for result in results:
        boxes = result.boxes
        if boxes is not None:
            for box in boxes:
                # Get box coordinates
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
                conf = box.conf[0].cpu().numpy()
                cls = int(box.cls[0].cpu().numpy())

                # Get class name
                class_name = model.names[cls]

                # Draw bounding box
                cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)

                # Draw label
                label = f"{class_name}: {conf:.2f}"
                cv2.putText(image_rgb, label, (x1, y1-10),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the image
    plt.figure(figsize=(12, 8))
    plt.imshow(image_rgb)
    plt.axis('off')
    plt.title('Helmet Detection Results')

    if save_path:
        plt.savefig(save_path, bbox_inches='tight', dpi=150)

    plt.show()

def batch_inference(model, image_folder, output_folder, conf_threshold=0.5):
    """
    Run inference on multiple images
    """
    os.makedirs(output_folder, exist_ok=True)

    image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
    image_files = [f for f in os.listdir(image_folder)
                   if any(f.lower().endswith(ext) for ext in image_extensions)]

    print(f"Processing {len(image_files)} images...")

    for image_file in image_files:
        image_path = os.path.join(image_folder, image_file)
        results = predict_image(model, image_path, conf_threshold)

        # Save annotated image
        annotated_image = results[0].plot()
        output_path = os.path.join(output_folder, f"detected_{image_file}")
        cv2.imwrite(output_path, annotated_image)

        print(f"Processed: {image_file}")

    print(f"All images processed. Results saved in: {output_folder}")

# =============================================================================
# 7. TEST THE MODEL
# =============================================================================

# Test on validation images
val_images_path = os.path.join(dataset_path, "valid", "images")
if os.path.exists(val_images_path):
    val_images = os.listdir(val_images_path)
    if val_images:
        # Test on first few validation images
        test_images = val_images[:5]  # Test on first 5 images

        print(f"Testing on {len(test_images)} validation images...")

        for img_name in test_images:
            img_path = os.path.join(val_images_path, img_name)
            print(f"\nTesting on: {img_name}")

            # Run prediction
            results = predict_image(best_model, img_path, conf_threshold=0.5)

            # Visualize results
            visualize_predictions(img_path, results)

            # Print detection details
            for result in results:
                boxes = result.boxes
                if boxes is not None:
                    print(f"  Detected {len(boxes)} objects")
                    for i, box in enumerate(boxes):
                        conf = box.conf[0].cpu().numpy()
                        cls = int(box.cls[0].cpu().numpy())
                        class_name = best_model.names[cls]
                        print(f"    Object {i+1}: {class_name} (confidence: {conf:.3f})")
                else:
                    print("  No objects detected")

# =============================================================================
# 8. EXPORT MODEL FOR DEPLOYMENT
# =============================================================================

# Export to different formats
print("Exporting model to different formats...")

# Export to ONNX (for deployment)
best_model.export(format='onnx')
print("Model exported to ONNX format")

# Export to TensorRT (if you have TensorRT installed)
# best_model.export(format='engine')

# Export to CoreML (for iOS)
# best_model.export(format='coreml')

# =============================================================================
# 9. UTILITY FUNCTIONS FOR DEPLOYMENT
# =============================================================================

def load_model_for_inference(model_path):
    """
    Load trained model for inference
    """
    model = YOLO(model_path)
    return model

def real_time_detection(model, source=0):
    """
    Real-time helmet detection using webcam
    """
    cap = cv2.VideoCapture(source)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Run inference
        results = model(frame, conf=0.5)

        # Visualize results
        annotated_frame = results[0].plot()

        # Display frame
        cv2.imshow('Helmet Detection', annotated_frame)

        # Break on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# =============================================================================
# 10. USAGE EXAMPLES
# =============================================================================

print("\n" + "="*50)
print("TRAINING COMPLETED SUCCESSFULLY!")
print("="*50)

print(f"\nBest model saved at: /content/helmet_detection/train_run/weights/best.pt")
print(f"Validation mAP50: {val_results.box.map50:.3f}")
print(f"Validation mAP50-95: {val_results.box.map:.3f}")

print("\nUsage Examples:")
print("1. Load model for inference:")
print("   model = load_model_for_inference('/content/helmet_detection/train_run/weights/best.pt')")

print("\n2. Single image prediction:")
print("   results = predict_image(model, 'path/to/image.jpg')")

print("\n3. Batch processing:")
print("   batch_inference(model, 'input_folder/', 'output_folder/')")

print("\n4. Real-time detection:")
print("   real_time_detection(model)  # Uses webcam")

print("\n5. Custom image prediction:")
print("   # Upload your image and run:")
print("   # results = predict_image(best_model, 'your_image.jpg')")
print("   # visualize_predictions('your_image.jpg', results)")

# =============================================================================
# 11. PERFORMANCE METRICS VISUALIZATION
# =============================================================================

def plot_training_metrics():
    """
    Plot training metrics if available
    """
    results_dir = "helmet_detection/train_run" # Corrected path

    # Check if results.png exists
    results_img = os.path.join(results_dir, "results.png")
    if os.path.exists(results_img):
        print("Training Results:")
        display(Image(results_img))

    # Check if confusion matrix exists
    confusion_matrix = os.path.join(results_dir, "confusion_matrix.png") # Corrected path
    if os.path.exists(confusion_matrix):
        print("Confusion Matrix:")
        display(Image(confusion_matrix))

# Display training results
plot_training_metrics()

print("\n" + "="*50)
print("HELMET DETECTION SETUP COMPLETE!")
print("You can now use the trained model for helmet detection.")
print("="*50)
```

```python
# Step 3: Create an Inference Function
def detect_helmets(image):
    """
    Runs helmet detection on an image and returns the annotated image.
    """
    # Ensure the input is a NumPy array (as expected by YOLO model inference)
    if not isinstance(image, np.ndarray):
        image = np.array(image)

    # Run inference
    results = model(image) # Use the loaded model

    # The results object contains the predictions.
    # The plot() method draws the bounding boxes and labels on the image.
    annotated_image = results[0].plot()

    return annotated_image
```

```python
# Step 4: Build and Launch Gradio Interface
import gradio as gr

# Create the Gradio interface
# The inputs will be an image, and the output will be an image
interface = gr.Interface(
    fn=detect_helmets,
    inputs=gr.Image(type="pil"), # Accept PIL Image as input
    outputs=gr.Image(type="numpy"), # Return numpy array for visualization
    title="Helmet Detection with YOLOv8",
    description="Upload an image to detect helmets, no-helmets, number plates, and riders."
)

# Launch the interface
# Setting share=True will create a public link for easy sharing
interface.launch(share=True)
```
